```python
csv_files = ["data/features.csv"]
xlsx_files = []

for csv_file in csv_files:
    df = pd.read_csv(csv_file)
    xlsx_file = csv_file.replace('.csv', '.xlsx')
    xlsx_files.append(xlsx_file)
    df.to_excel(xlsx_file, index=False)
```
